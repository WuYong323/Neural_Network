```python
import torch
```

```python
z=torch.zeros(3,2)
print(z)
print(z.dtype)
```

```python
#随机数主要用于实验复现
torch.manual_seed(1724)
r1=torch.rand(2,3)
print(r1)

r2=torch.rand(2,3)
print(r2)

torch.manual_seed(1729)
r3=torch.rand(2,3)
print(r3)
```

```python
one=torch.ones(2,3)
print(one)
two=torch.ones(2,3)
print(two)
three=one+two
print(three)
print(three.dtype)
```

```python
r=torch.rand(2,2)*2-1
print(r)
print(torch.abs(r))
print(torch.asin(r))
print(torch.std_mean(r))
print(torch.max(r))
```

```python
a=torch.tensor([[1.,2.],
                [3.,4.],
                [5.,6.]])
u,s,v=torch.svd(a)
print(u)
print(s)
print(v)
```
